```python
df = df.drop(['closed_at','id','latitude','longitude','Unnamed: 6','name','state_code.1','labels','object_id','age_first_milestone_year','age_last_milestone_year'],axis =1)
```

```python
cate = df.select_dtypes(include = 'object').columns
cate
```

```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
for i in cate:
  df[i] = lbl.fit_transform(df[i])
```

```python
df.info()
```

```python
plt.figure(figsize = (20,20))
crr = df.corr()
sns.heatmap(crr,annot = True,linewidths= .5)
```

```python
X =df.drop('status',axis =1)
Y = df['status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y)
```

```python
from sklearn.feature_selection import RFE
from sklearn.naive_bayes import BernoulliNB
gnb = BernoulliNB()
rfe = RFE(gnb,n_features_to_select=10)
```

```python
gnb.fit(x_train, y_train)
```

```python
y_pred = gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
